```python
# Lista de Archivos de la carpeta
ruta_principal = '/content/drive/MyDrive/Colab Notebooks/'
ruta_no_etiquetados = ruta_principal + 'DatosNoEtiquetados.xlsx'
ruta_etiquetados = ruta_principal + 'DatosEtiquetados.xlsx'
ruta_etiquetados2 = ruta_principal + 'DatosEtiquetados_v2.xlsx'

ruta_etiquetados3 = ruta_principal + 'DatosEtiquetadosV3.xlsx'
```

```python
# Importar Archivo Excel
df_labeled_data = pd.read_excel(ruta_etiquetados3)
labeled_data = np.array(df_labeled_data)
labeled_data
```

```python
# Importar Archivo Excel
df_unlabeled_data = pd.read_excel(ruta_no_etiquetados)
#unlabeled_data = np.array(df_unlabeled_data)
#unlabeled_data = df_unlabeled_data.values.tolist()
unlabeled_data = [str(data) for data in df_unlabeled_data.iloc[:, 0]]
unlabeled_data
```

```python
# Dividir los datos etiquetados en entrenamiento y prueba
X_labeled = [data[0] for data in labeled_data]
y_labeled = [data[1] for data in labeled_data]
X_train, X_test, y_train, y_test = train_test_split(X_labeled, y_labeled, test_size=0.2, random_state=42)

# Preprocesamiento de texto y clasificación con Naive Bayes
stop_words = stopwords.words('spanish')
pipeline = Pipeline([
    ('vectorizer', TfidfVectorizer(stop_words=stop_words)),
    ('classifier', MultinomialNB()),
])
```

```python
# Entrenar el modelo
pipeline.fit(X_train, y_train)
```

```python
# Guardar el modelo entrenado en un archivo
joblib.dump(pipeline, 'modelo_entrenado.pkl')
```

```python
# Evaluar el modelo
y_pred = pipeline.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Precisión del modelo:", accuracy)
```
